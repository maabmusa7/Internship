## Step 1 :
This function takes a list of numbers as an argument and finds the mean, minumum and maximum using predefined python funtions (sum(), min(), max()).
It returns the values in the form of a dictionary.

```python
#step 1
def stats(nums):
    """The mean, min and max of your list are:"""
    return {
        "mean": sum(nums) / len(nums),
        "min": min(nums),
        "max": max(nums)
    }

  
```

```python
print(stats([5,2,8,13,3,7,-5]))
```

## Step 2 :
This list comprehension is an alternative for a for loop that lists the even numbers from a list of numbers.

```python
#step 2
evens = [x for x in range(10) if x % 2 == 0 ]
```

```python
print(evens) 
```

## Step 3 : 
This class organises Movies by their title and genre (attributes) and provides a method for accessing movie details.

```python
#step 3
class Movie:
    def __init__(self, title, genre):
        self.title = title
        self.genre = genre

    def get_movie_details(self):
        return "Title: " + self.title +", Genre: " + self.genre
```

```python
M = Movie("Inception", "Sci-Fi")
print(M.get_movie_details())
```
