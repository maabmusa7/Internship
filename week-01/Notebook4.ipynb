## Step 1:
- Pandas library was imported first.
- read_csv() used to convert the dataset into a DataFrame.
- A total of 891 rows and 12 columns are availabe.

```python
#step1
import pandas as pd
df= pd.read_csv("train.csv")

print(df.shape)
```

```python
columns = df.columns

print(columns)
```

"columns" used to list the dataset column names and print them out.

```python
print(df.dtypes)
```

"dtypes" used to list the data type of each column.

## Step 2: 

```python
#step2 
missing = df.isnull().sum()

print(missing)
```

Rows with missing values are listed above.

```python
df_new = df.drop(columns=["Cabin"])
print(df_new)
```

The DataFrame is modified by dropping the "Cabin" column - has a lot of missing rows- because deleting the missing rows entirely will remove most of the table data.

```python
x = df_new["Age"].median()
print("Average age:", x)
df_new["Age"]= df_new["Age"].fillna(x)
print(df_new)
```

"Age" column, in contrast, is filled on the missing rows with the average age for efficiency.

```python
df_new = df_new.dropna(subset=["Embarked"])
print(df_new)
```

Rows with missing data in "Embarked" column are dropped (minimal impact), bringing the table to a total of 889 rows and 11 columns.

```python
df_new = df_new.drop_duplicates()
print(df_new.shape)
```

Since there is no duplicates, the number of rows remains the same.

## Step 3:

```python
#step3
df_avg_fare = df_new["Fare"].mean()
print(df_avg_fare)
```

The average Fare calculated using mean() and printed.

```python
high_fare = df_new[df_new["Fare"]>df_avg_fare]
print(high_fare.shape[0], "out of", df_new.shape[0],"passengers paid above the average!")
print(high_fare)
```

Fares higher than the average are filterd with boolean masking and printed.

### Findings:
  1- The average Fare for is 32.09
  2- Fares higher than the average are 209 out of 889 (approximately quarter the total Fares).
  3- Mjority of passengers who paid above the average are first class passengers.

# Step 4:
- "groupby" and "agg()"used to group data to display survival rate, Age and Fare per class.
- It was also used to calculate the survival rate based on Gender.

```python
#step4
class_based = df_new.groupby("Pclass").agg({"PassengerId" : "count",
                             "Survived" : "mean",
                             "Age" : "mean",
                             "Fare" : "mean"}
                            )
print(class_based)
```

Fisrt class passengers have the highest survival rate while paying the highest Fares in Average.
Third class passengers form a majority of the train passengers and they form the youngest group of passengers in average.

```python
gender_based = df_new.groupby("Sex")["Survived"].mean()   
print(gender_based)
```

Most survivors of the train are females.
