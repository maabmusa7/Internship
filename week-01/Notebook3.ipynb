## Step 3:
A 2D 3x3 Numpy array.
mean() function used for calculating the mean of the "arr" array and printing values greater than it.

```python
#step3
arr = np.array([
    [1,16,30],
    [25,80,42],
    [7,22,50]
])
```

```python
print("Array mean:",arr.mean())
print("Values greater than array mean:", arr[arr > arr.mean()])
```

# Step 4:
A 2D 3x5 Numpy array and a 1D 1x5 array.
since both arrays have 5 columns, it allows broadcasting.
A manual and a broadcasted result show consistency when compaired using the == operator.

```python
#step4

twoD = np.array([
    [1,0,5,7,2],
    [8,4,6,2,9],
    [7,3,10,8,1],
])

oneD = np.array([1,6,3,5,2])

```

```python
print(twoD.shape)
print(oneD.shape)
```

```python
oneD_replicated = np.array(
                [[1,6,3,5,2],
                 [1,6,3,5,2],
                 [1,6,3,5,2]])
```

```python
manual_result = twoD + oneD_replicated
print(manual_result)
```

```python
broadcast_result= twoD + oneD
print(broadcast_result)
```

```python
manual_result == broadcast_result
```
